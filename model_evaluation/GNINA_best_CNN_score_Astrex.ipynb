```python
# ==== GNINA robust batch docking (Colab-safe) =================================
import os, re, sys, shlex, time, json, shutil, subprocess, tempfile, textwrap
import pandas as pd
from pathlib import Path

# ----------------- USER INPUTS -----------------
BASE_DIR    = '/content/gdrive/MyDrive/Docking_benchmarks/PB_new_run/astex_diverse_set_Ensemble_run_7'
LIST_FILE   = '/content/gdrive/MyDrive/Docking_benchmarks/PB_new_run/astex_diverse_set_ids.txt'
GNINA_BIN   = '/content/gnina'       # wget ...; mv gnina.1.3.2 gnina; chmod +x gnina
CPU_CORES   = os.cpu_count()
EXHAUST     = 32
NUM_MODES   = 40
CNN_SCORING = 'rescore'              # 'rescore' recommended; 'all' slower; 'vina' disables CNN
VERBOSITY   = '2'
GPU_DEVICE  = '0'                    # '0' for first GPU; use 'cpu' to force CPU
FAST_MODE   = True                   # quick pass uses lower exha/modes
SPLIT_TO_SDF= True                   # requires OpenBabel in PATH: obabel
RANK_BY     = "cnn"                  # "cnn" (default) or "vina"

# thresholds & pocket filters
THRESHOLDS_ALL = [2, 5, 10, 15]
POCKET_MATCH   = ("pocket_thr05_Minimal", "pocket_thr05_Majority")  # file name contains

# ----------------- UTILITIES -----------------
def ensure_dir(p: str):
    Path(p).mkdir(parents=True, exist_ok=True)

def check_bin(path: str, name="binary"):
    if not Path(path).exists():
        raise FileNotFoundError(f"{name} not found: {path}")
    if not os.access(path, os.X_OK):
        raise PermissionError(f"{name} is not executable: {path}")

def run_cmd(cmd_list, log_path=None, env=None, cwd=None, timeout=None):
    """
    Run a command, streaming stdout/stderr to both cell output and a log file.
    Returns (exit_code, elapsed_seconds).
    """
    start = time.time()
    lf = None
    try:
        if log_path:
            ensure_dir(os.path.dirname(log_path))
            lf = open(log_path, 'w', buffering=1, encoding='utf-8', errors='ignore')
        proc = subprocess.Popen(
            cmd_list,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            cwd=cwd,
            env=env,
            text=True,
            bufsize=1,
        )
        for line in proc.stdout:
            line = line.rstrip('\n')
            print(line)
            if lf: lf.write(line + '\n')
        proc.wait(timeout=timeout)
        code = proc.returncode
    except subprocess.TimeoutExpired:
        code = 124
    except FileNotFoundError:
        code = 127
    except Exception as e:
        code = 1
        msg = f"[run_cmd EXCEPTION] {e}"
        print(msg)
        if lf: lf.write(msg + '\n')
    finally:
        if lf: lf.close()
    return code, time.time() - start

# remark patterns seen in GNINA PDBQT
REMARK_RE_1 = re.compile(r"REMARK\s+(\w+)\s*=\s*([-\d\.eE+]+)")
REMARK_RE_2 = re.compile(r"REMARK\s+(\w+)\s+([-\d\.eE+]+)")

def parse_gnina_pdbqt(pdbqt_path: str) -> pd.DataFrame:
    """
    Parse GNINA PDBQT for per-model REMARKs:
      - multiple MODEL/ENDMDL blocks with REMARK keys (minimizedAffinity, CNNscore, CNNaffinity, affinity, intramol, etc.)
      - fallback: REMARKs but no MODEL -> treat as single MODEL=1
    Returns empty DF if no models.
    """
    if not os.path.exists(pdbqt_path) or os.path.getsize(pdbqt_path) == 0:
        return pd.DataFrame()

    rows, cur, in_model = [], {}, False
    whole = {}
    saw_any_remark = False

    with open(pdbqt_path, 'r', errors='ignore') as f:
        for raw in f:
            line = raw.strip()
            if line.startswith("MODEL"):
                in_model = True
                if cur: rows.append(cur)
                parts = line.split()
                try:
                    model_n = int(parts[1])
                except:
                    model_n = (rows[-1]["MODEL"]+1 if rows and "MODEL" in rows[-1] else 1)
                cur = {"MODEL": model_n}
            elif line.startswith("REMARK"):
                m = REMARK_RE_1.match(line) or REMARK_RE_2.match(line)
                if m:
                    k, v = m.groups()
                    try: v = float(v)
                    except: pass
                    if in_model:
                        cur[k] = v
                    else:
                        whole[k] = v
                    saw_any_remark = True
            elif line.startswith("ENDMDL"):
                if cur: rows.append(cur)
                cur, in_model = {}, False

    if rows:
        df = pd.DataFrame(rows)
    elif saw_any_remark:
        one = {"MODEL": 1}; one.update(whole)
        df = pd.DataFrame([one])
    else:
        df = pd.DataFrame()

    return normalize_columns(df)

def normalize_columns(df: pd.DataFrame) -> pd.DataFrame:
    """
    Normalize common key variants to a consistent set:
      minimizedAffinity, CNNscore, CNNaffinity, affinity, intramol
    """
    if df.empty:
        return df
    rename = {}
    # case/variant normalizations
    for c in list(df.columns):
        lc = c.lower()
        if lc == "minimized":
            rename[c] = "minimizedAffinity"
        elif lc in ("minimizedaffinity",):
            rename[c] = "minimizedAffinity"
        elif lc in ("cnnscore", "cnn_score"):
            rename[c] = "CNNscore"
        elif lc in ("cnnaffinity", "cnn_affinity"):
            rename[c] = "CNNaffinity"
        elif lc in ("vina", "affinity_kcal_mol", "affinity"):
            rename[c] = "affinity"
        elif lc in ("intramol", "intramolecular"):
            rename[c] = "intramol"
    return df.rename(columns=rename)

def rank_models(df: pd.DataFrame, by="cnn") -> pd.DataFrame:
    """
    Sort models so the first row is the 'best':
      by="cnn":   highest CNNscore, then highest CNNaffinity (pK), then most negative minimizedAffinity, then affinity
      by="vina":  most negative minimizedAffinity, then affinity
    Stable mergesort keeps original order among ties.
    """
    df = normalize_columns(df.copy())
    cols = df.columns

    if by == "vina":
        keys = []
        if "minimizedAffinity" in cols: keys.append(("minimizedAffinity", True))  # ascending=True (more negative better)
        if "affinity" in cols:          keys.append(("affinity", True))
        if not keys:
            return df.reset_index(drop=True)
        by_cols = [k for k,_ in keys]
        asc    = [True for _ in keys]
        return df.sort_values(by=by_cols, ascending=asc, kind='mergesort').reset_index(drop=True)

    # default: CNN
    keys = []
    if "CNNscore" in cols:           keys.append(("CNNscore", False))       # higher is better
    if "CNNaffinity" in cols:        keys.append(("CNNaffinity", False))    # pK units: higher is better
    if "minimizedAffinity" in cols:  keys.append(("minimizedAffinity", True))# more negative better
    if "affinity" in cols:           keys.append(("affinity", True))
    if not keys:
        return df.reset_index(drop=True)

    by_cols = [k for k,_ in keys]
    asc     = [a for _,a in keys]
    return df.sort_values(by=by_cols, ascending=asc, kind='mergesort').reset_index(drop=True)

def write_best_first(in_pdbqt: str, out_pdbqt: str, best_model: int) -> bool:
    """
    Reorder models so that best_model becomes MODEL 1 and is written first.
    For single-model files, just copy.
    """
    try:
        lines = Path(in_pdbqt).read_text(errors='ignore').splitlines()
    except Exception:
        return False

    # Split into MODEL blocks
    blocks, cur, cur_id = [], [], None
    for ln in lines:
        if ln.startswith("MODEL"):
            if cur: blocks.append((cur_id, cur)); cur=[]
            parts = ln.split()
            try: cur_id = int(parts[1])
            except: cur_id = None
            cur = [ln]
        elif ln.startswith("ENDMDL"):
            cur.append(ln)
            blocks.append((cur_id, cur))
            cur, cur_id = [], None
        else:
            cur.append(ln)

    if blocks:
        # reorder: best first as MODEL 1, renumber sequentially
        idx = next((i for i,(mid,_) in enumerate(blocks) if mid == best_model), 0)
        reordered = [blocks[idx]] + [b for i,b in enumerate(blocks) if i != idx]
        out_lines = []
        for i,(_, blk) in enumerate(reordered, start=1):
            for ln in blk:
                if ln.startswith("MODEL"):
                    out_lines.append(f"MODEL        {i}")
                else:
                    out_lines.append(ln)
        Path(out_pdbqt).write_text("\n".join(out_lines))
        return True

    # No explicit MODEL blocks ‚Äî just copy over (treat as single model)
    shutil.copy2(in_pdbqt, out_pdbqt)
    return True

def dump_csv(df_all: pd.DataFrame, best_row: pd.Series, csv_all: str, csv_best: str, job_id: str):
    ensure_dir(os.path.dirname(csv_all))
    df_all2 = normalize_columns(df_all).copy()
    df_all2.insert(0, "ID", job_id)
    df_all2.to_csv(csv_all, index=False)
    # make sure best row has normalized columns
    best_norm = normalize_columns(pd.DataFrame([best_row])).iloc[0]
    pd.DataFrame([best_norm]).assign(ID=job_id).to_csv(csv_best, index=False)

def split_to_sdfs(pdbqt_path: str, base_out: str):
    if not SPLIT_TO_SDF: return True
    if shutil.which("obabel") is None:
        print("[WARN] obabel not found. Skipping SDF split.")
        return False
    # -m => separate files: <base_out>1.sdf, <base_out>2.sdf, ...
    cmd = ["obabel", "-ipdbqt", pdbqt_path, "-osdf", "-m", "-O", f"{base_out}.sdf"]
    code, _ = run_cmd(cmd)
    return code == 0

def has_valid_existing(docking_folder: str, out_pdbqt: str) -> bool:
    """
    Skip condition you asked: one valid _output.pdbqt AND at least one .sdf.
    We also verify the pdbqt parses to >=1 model.
    """
    if not (os.path.exists(out_pdbqt) and os.path.getsize(out_pdbqt) > 0):
        return False
    any_sdf = any(str(p).lower().endswith(".sdf") for p in Path(docking_folder).glob("*.sdf"))
    if not any_sdf:
        return False
    df = parse_gnina_pdbqt(out_pdbqt)
    return not df.empty

# ----------------- MAIN -----------------
# sanity checks
check_bin(GNINA_BIN, "gnina")

# Read all subfolders to process
with open(LIST_FILE, 'r') as fh:
    targets = [ln.strip() for ln in fh if ln.strip()]

# scratch (fast local IO to avoid Drive flakiness)
SCRATCH = Path("/content/_gnina_scratch")
ensure_dir(SCRATCH)

# global CSV of bests across pockets (append mode)
GLOBAL_BEST_CSV = Path("/content/_gnina_scratch/_ALL_BESTS.csv")
if not GLOBAL_BEST_CSV.exists():
    GLOBAL_BEST_CSV.write_text("ID,pocket,threshold,MODEL,CNNscore,CNNaffinity,minimizedAffinity,affinity\n")

jobs_total = 0
jobs_done  = 0

for sub in targets:
    sub_dir = Path(BASE_DIR) / sub
    if not sub_dir.is_dir():
        print(f"[SKIP] Missing dir: {sub_dir}")
        continue

    for pocket_file in os.listdir(sub_dir):
        if not pocket_file.endswith(".pdb"):
            continue
        if not any(tok in pocket_file for tok in POCKET_MATCH):
            continue

        pocket_name = os.path.splitext(pocket_file)[0]
        for thr in THRESHOLDS_ALL:
            # For Majority pockets, only 2 and 5
            if "pocket_thr05_Majority" in pocket_file and thr not in (2, 5):
                continue

            protein_file = sub_dir / f"{os.path.basename(sub)}_protein_reduce_rm_bad_input.pdbqt"
            ligand_file  = sub_dir / f"{os.path.basename(sub)}_ligand_start_conf_prepared.pdbqt"
            config_file  = sub_dir / f"{pocket_name}_config_{thr}.txt"
            docking_dir  = sub_dir / f"{pocket_name}_results_{thr}"
            ensure_dir(docking_dir)

            ID = f"{os.path.basename(sub)}_{pocket_name}_{thr}"

            out_pdbqt = docking_dir / f"{ID}_output.pdbqt"
            out_log   = docking_dir / f"{ID}_output.log"
            csv_all   = docking_dir / f"{ID}_poses.csv"
            csv_best  = docking_dir / f"{ID}_best.csv"

            s_dir   = SCRATCH / ID
            ensure_dir(s_dir)
            s_pdbqt = s_dir / f"{ID}_output.pdbqt"
            s_log   = s_dir / f"{ID}_output.log"

            jobs_total += 1
            # --- Preflight ---
            bad = []
            for p,desc in [(protein_file,"protein pdbqt"), (ligand_file,"ligand pdbqt"), (config_file,"config txt")]:
                if not Path(p).exists() or os.path.getsize(p) == 0:
                    bad.append(f"{desc} missing/empty: {p}")
            if bad:
                msg = f"[FAIL: preflight] {ID} ‚Äî " + " | ".join(bad)
                print(msg)
                (docking_dir / "FAIL.txt").write_text(msg + "\n")
                continue

            # --- Smart skip/resume ---
            if has_valid_existing(str(docking_dir), str(out_pdbqt)):
                print(f"[SKIP] {ID} ‚Äî valid output+SDF already present.")
                jobs_done += 1
                continue

            # POST-only recovery: if out_pdbqt exists but no SDF, reuse it
            reuse_ok = False
            if out_pdbqt.exists() and out_pdbqt.stat().st_size > 0:
                df_probe = parse_gnina_pdbqt(str(out_pdbqt))
                if not df_probe.empty:
                    shutil.copy2(out_pdbqt, s_pdbqt)
                    reuse_ok = True
                    print(f"[POST only] {ID} ‚Äî reusing prior _output.pdbqt for post-processing.")

            # --- Dock if needed ---
            if not reuse_ok:
                print(f"\n=== Starting docking for {ID} (FAST={FAST_MODE}) ===")
                exha = 32
                num_m = 40

                cmd = [
                    GNINA_BIN,
                    "--receptor", str(protein_file),
                    "--ligand",   str(ligand_file),
                    "--out",      str(s_pdbqt),
                    "--config",   str(config_file),
                    "--cpu",      str(CPU_CORES),
                    "--exhaustiveness", str(exha),
                    "--num_modes", str(num_m),
                    "--cnn_scoring", CNN_SCORING,
                    "--verbosity", VERBOSITY
                ]
                env = os.environ.copy()
                env["OMP_NUM_THREADS"] = "1"
                env["MKL_NUM_THREADS"] = "1"
                env["OPENBLAS_NUM_THREADS"] = "1"
                if GPU_DEVICE != "cpu":
                    env["CUDA_VISIBLE_DEVICES"] = env.get("CUDA_VISIBLE_DEVICES", GPU_DEVICE)

                code, elapsed = run_cmd(cmd, log_path=str(s_log), env=env)
                print(f"Docking completed for {ID} in {elapsed:.2f} s (exit={code}).")
                if code != 0 or not (s_pdbqt.exists() and s_pdbqt.stat().st_size > 0):
                    tail = ""
                    if s_log.exists():
                        try:
                            tail = "\n".join(s_log.read_text(errors='ignore').splitlines()[-60:])
                        except: pass
                    msg = f"[FAIL] {ID} ‚Äî gnina did not produce output. Log tail:\n{tail if tail else '[no log]'}"
                    print(msg)
                    (docking_dir / "FAIL.txt").write_text(msg + "\n")
                    if s_log.exists():
                        shutil.copy2(s_log, out_log)
                    continue

            # --- Postprocess: parse, rank, reorder MODEL 1, CSVs, SDFs ---
            df = parse_gnina_pdbqt(str(s_pdbqt))
            if df.empty:
                msg = f"[POST ERR] {ID}: No models parsed from {s_pdbqt}"
                print(msg)
                (docking_dir / "FAIL.txt").write_text(msg + "\n")
                if s_log.exists():
                    shutil.copy2(s_log, out_log)
                continue

            df_ranked = rank_models(df, by=RANK_BY)
            best = df_ranked.iloc[0].copy()
            best_model = int(best.get("MODEL", 1))

            # reorder so best becomes MODEL 1 at the top
            s_pdbqt_best = s_dir / f"{ID}_reordered.pdbqt"
            ok = write_best_first(str(s_pdbqt), str(s_pdbqt_best), best_model=best_model)
            final_pdbqt = s_pdbqt_best if ok else s_pdbqt

            # write CSVs
            dump_csv(df_ranked, best, str(s_dir / "poses.csv"), str(s_dir / "best.csv"), job_id=ID)

            # split to SDFs
            sdf_ok = split_to_sdfs(str(final_pdbqt), str(docking_dir / f"{ID}_pose"))

            # move artifacts to docking folder
            try:
                shutil.copy2(final_pdbqt, out_pdbqt)
                shutil.copy2(s_dir / "poses.csv", csv_all)
                shutil.copy2(s_dir / "best.csv",  csv_best)
                if s_log.exists():
                    shutil.copy2(s_log, out_log)
            except Exception as e:
                print(f"[WARN] Copy-out failed for {ID}: {e}")

            # append to global bests
            best_norm = normalize_columns(pd.DataFrame([best])).iloc[0]
            try:
                with open(GLOBAL_BEST_CSV, "a") as gh:
                    gh.write(f"{ID},{pocket_name},{thr},{int(best_norm.get('MODEL',1))},"
                             f"{best_norm.get('CNNscore','')},{best_norm.get('CNNaffinity','')},"
                             f"{best_norm.get('minimizedAffinity','')},{best_norm.get('affinity','')}\n")
            except Exception as e:
                print(f"[WARN] Could not append to global csv: {e}")

            # final report
            print(f"[{ID}] Best ({'CNN' if RANK_BY=='cnn' else 'Vina'}) ‚Üí "
                  f"MODEL={best_model} | "
                  f"CNNscore={best_norm.get('CNNscore','?')} | "
                  f"CNNaffinity={best_norm.get('CNNaffinity','?')} | "
                  f"Vina(min)={best_norm.get('minimizedAffinity', best_norm.get('affinity','?'))}")
            jobs_done += 1

print(f"\nDone. {jobs_done}/{jobs_total} jobs finished or skipped successfully.")
print(f"Global bests: {GLOBAL_BEST_CSV}")
```
